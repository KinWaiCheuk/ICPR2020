```python
import torch
import torch.nn as nn
```

```python
class BiLSTM(nn.Module):
    inference_chunk_length = 512

    def __init__(self, input_features, recurrent_features):
        super().__init__()
        self.rnn = nn.LSTM(input_features, recurrent_features, batch_first=True, bidirectional=True)

    def forward(self, x):
        return self.rnn(x)
```

```python
x = torch.randn(8, 640, 768)
```

```python
bilstm_layer = BiLSTM(768,768//2)
```

```python
output, (h_n, c_n) = bilstm_layer(x)
```
